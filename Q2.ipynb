```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("p2p-Gnutella06.txt", sep = '	', header = None)
data
```

```python
if data[0].max() >= data[1].max():
    nodes = data[0].max()
else:
    nodes = data[1].max()
nodes += 1
edges = data.shape[0]
print("Number of nodes: ", nodes)
print("Number of edges: ", edges)
```

```python
#Adjacency list for outgoing edges
adj_out = []
for i in range(nodes):
    adj_out.append(list(data[data[0] == i][1]))
adj_out
```

```python
#Adjacency list for incoming edges
adj_in = []
for i in range(nodes):
    adj_in.append(list(data[data[1] == i][0]))
adj_in
```

## PageRank

```python
prscores = np.ones(nodes)
prscores = prscores/nodes
prscores
```

```python
df = 0.85
itr = 100
for i in range(itr):
    for j in range(nodes):
        prscores[j] = 1 - df
        for k in adj_in[j]:
            prscores[j] += ( df * (prscores[k]/len(adj_out[k])))
```

```python
prscores
```

## Authority and Hub

```python
hub = np.ones(nodes)
auth = np.ones(nodes)
hub, auth
```

```python
itr = 100
for i in range(itr):
    hub_sum = 0
    auth_sum = 0
    for j in range(nodes):
        temp = 0 
        for k in adj_in[j]:
            temp += hub[k]
        auth[j] = temp
        auth_sum += temp
        temp = 0
        for k in adj_out[j]:
            temp += auth[k]
        hub[j] = temp
        hub_sum += temp
    for j in range(nodes):
        hub[j] /= hub_sum
        auth[j] /= auth_sum
```

```python
hub
```

```python
auth
```

## Results

```python
#Node having max prscore
best_prs = np.argmax(prscores)
print("Node having max prscore: ", best_prs, "({})".format(prscores[best_prs]))
```

```python
#Node having max hub score
best_hub = np.argmax(hub)
print("Node having max Hub score: ", best_hub, "({})".format(hub[best_hub]))
```

```python
#Node having max auth score
best_auth = np.argmax(auth)
print("Node having max auth score: ", best_auth, "({})".format(auth[best_auth]))
```

```python
scores = pd.DataFrame({"Node" : range(nodes), "PageRank" : prscores, "Hub": hub, "Authority" : auth})
scores
```

```python
pr_df = scores.sort_values('PageRank', ascending = False)
pr_df
```
